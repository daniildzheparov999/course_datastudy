Series – структура данных в pandas, которая представляет собой одномерный массив. Аналогом такого одномерного массива в python являются списки. 
В списке можно обращаться к элементу по его индексу, который всегда является порядковым номером элемента. В отличие от списка, в Series индексами могут быть не только порядковые номера, но что угодно другое, например даты, названия компаний, идентификаторы участников опроса и т.д.


```python
#создаем одномерный массив с индесами (Series) и выводим его на экран с помощью функции print()
test_series = pd.Series(['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'], 
                        index = ['1 day', '2 day', '3 day', '4 day', '5 day','6 day', '7 day'])

print(test_series)
```

```python
#обращаемся к значению в массиве через идекс
test_series['1 day']
```

```python
#для сравнения создаем обычный список в python. индексы здесь будут уже числовые и начинаются с 0
python_series = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
print(python_series)
```

```python
#образаемся к эелементу списка через индекс
python_series[0]
```

DataFrame – структура данных в pandas, которую можно представлять в виде таблицы. Есть атрибуты (столбцы) и значения (строки). Столбцы являются объектами Series, элементы которых составляют строки в DataFrame.

```python
#создание датафрейма через словарь (по столбцам). Задаем название столбца и потом перечисляем все его значения
df = pd.DataFrame({'Name of day': ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday'], 
                    'Type of day': ['working', 'working', 'working', 'working', 'working', 'holiday', 'holiday'] })

df
```

```python
#создаем датафрейм через списки (по строкам)
df1 = pd.DataFrame([['monday', 'working'], 
                    ['tuesday', 'working'],
                    ['wednesday', 'working'],
                    ['thursday', 'working'],
                    ['friday', 'working'],
                    ['saturday', 'holiday'],
                    ['sunday', 'holiday']],
                    columns = ['Name of day', 'Type of day'])

df1
```

Pandas позволяет считывать различные форматы файлов, преобразовывать из в датафрейм и дальше работать с данными. Можно работатьс файлами csv, xls, xlsx, json и др. 
Для файлов формата csv есть дополнительные параметры, которые помогают считывать данные правильно
sep — разделитель данных, по умолчанию ',';
decimal — разделитель числа на целую и дробную часть, по умолчанию'.';
names — список с названиями колонок (не обязательный параметр);
skiprows — если файл содержит ненужные строки сверху, которые нужно пропустить (не обязательный параметр).

select region_id, avg (transaction_amount) as avg_amount
from sales 
group by region_id
having avg_amount > 50000
order by avg_amount;

```python
sales = sales.groupby('region_id').agg({'transaction_amount': np.mean})
sales[sales['transaction_amount'] > 50000]
```

```python
sales = pd.read_csv('sales.csv')
sales_products = pd.merge(sales, products,  on = 'product_id')
sales_products
```

```python
sales_products_regions = pd.merge(sales_products, regions_df, on = 'region_id')
sales_products_regions
```

```python
final_df = pd.merge(sales_products_regions, time, on = 'time_id')
final_df
```

```python
final_df_for_export = final_df[['product_name', 'time_datetime', 'region_name', 'transaction_amount']]
final_df_for_export['time_datetime'] = final_df_for_export['time_datetime'].apply(pd.to_datetime)

final_df_for_export.info()
```

```python
#добавление поля с условием
final_df_for_export.loc[final_df_for_export['transaction_amount'] > 0, 'transaction_flag'] = 'positive'
final_df_for_export.loc[final_df_for_export['transaction_amount'] < 0, 'transaction_flag'] = 'negative'
final_df_for_export
```

```python
#добавляем сложные условия и применяем к новому полю
conditions = [
    (final_df_for_export['transaction_amount'] < 0),
    (final_df_for_export['transaction_amount'] > 0) & (final_df_for_export['transaction_amount'] < 30000),
    (final_df_for_export['transaction_amount'] > 30000) & (final_df_for_export['transaction_amount'] < 50000),
    (final_df_for_export['transaction_amount'] > 50000)
]

values = [
    'loss',
    'minimal',
    'medium',
    'high'
]

final_df_for_export['transaction__amount_category'] = np.select(conditions, values)

final_df_for_export
```

```python
#экспортируем файлы в формате xlsx
final_df_for_export.to_excel('final_df.xlsx')
```
